```python
def log_info(log):
    summary = {
        'Epoch': [],
        'Train unlabeled top@1': [],
        'Train unlabeled ece': []
    }

    with open(log, 'r') as log_file:
        for line in log_file:
            if line.startswith('2023-') and "Epoch:" in line:
                summary['Train unlabeled ece'].append(float(line[line.find('train_unlabeled_ece'):].split('|')[0].split(":")[-1]))
                summary['Train unlabeled top@1'].append(float(line[line.find('train_unlabeled_top@1'):].split('|')[0].split(":")[-1]))
                summary['Epoch'].append(int(line[line.find('Epoch'):][:line[line.find('Epoch'):].find("/100]")].split('[')[-1]))
    
    return summary

LOGS = {
    
    'CIFAR10': {'FIXMATCH':{'0.3':log_info('/mnthdd/Dropbox/D/personal_study/SafeSSL+Calibration/checkpoints/cifar10/FIXMATCH/wide28_2/2023-09-13_18-26-13/main.log'),
                            '0.6':log_info('/mnthdd/Dropbox/D/personal_study/SafeSSL+Calibration/checkpoints/cifar10/FIXMATCH/wide28_2/2023-09-14_04-52-33/main.log')},

                'CALIMATCH':{'0.3':log_info('/mnthdd/Dropbox/D/personal_study/SafeSSL+Calibration/checkpoints/cifar10/CaliMATCH/wide28_2/2023-09-17_18-33-41/main.log'),
                             '0.6':log_info('/mnthdd/Dropbox/D/personal_study/SafeSSL+Calibration/checkpoints/cifar10/CaliMATCH/wide28_2/2023-09-18_01-02-35/main.log')
                             }
                
                
                },
    
    'CIFAR100': {'FIXMATCH':{'0.3':log_info('/mnthdd/Dropbox/D/personal_study/SafeSSL+Calibration/checkpoints/cifar100/FIXMATCH/wide28_2/2023-09-13_23-37-42/main.log'),
                             '0.6':log_info('/mnthdd/Dropbox/D/personal_study/SafeSSL+Calibration/checkpoints/cifar100/FIXMATCH/wide28_2/2023-09-14_10-55-41/main.log')},
                 
                 'CALIMATCH':{'0.3':log_info('/mnthdd/Dropbox/D/personal_study/SafeSSL+Calibration/checkpoints/cifar100/CaliMATCH/wide28_2/2023-09-17_21-47-57/main.log'),
                              '0.6':log_info('/mnthdd/Dropbox/D/personal_study/SafeSSL+Calibration/checkpoints/cifar100/CaliMATCH/wide28_2/2023-09-18_04-16-53/main.log')}
                 
                 
                 }
    
}
```
