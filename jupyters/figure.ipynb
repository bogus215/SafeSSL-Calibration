```python
import pandas as pd
calimatch_cifar100 = pd.read_csv("CaliMatch+CIFAR100.csv").iloc[:,:2]
calimatch_cifar100 = calimatch_cifar100.rename(columns={calimatch_cifar100.columns[1]:"unlabeled_train_acc"})

fixmatch_cifar100 = pd.read_csv("FixMatch+CIFAR100.csv").iloc[:,:2]
fixmatch_cifar100 = fixmatch_cifar100.rename(columns={fixmatch_cifar100.columns[1]:"unlabeled_train_acc"})

import matplotlib.pyplot as plt
plt.style.use('bmh')
plt.plot(fixmatch_cifar100['Step'],fixmatch_cifar100['unlabeled_train_acc'], label='FixMatch')
plt.plot(calimatch_cifar100['Step'],calimatch_cifar100['unlabeled_train_acc'], label='Proposed (FixMatch+AcatS)')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('bmh')

for ratio in ["0.50" , "0.00"]:
    calimatch_cifar10 = pd.read_csv(f"CaliMatch+CIFAR10+ratio+{ratio}.csv").iloc[:,:2]
    calimatch_cifar10 = calimatch_cifar10.rename(columns={calimatch_cifar10.columns[1]:"unlabeled_train_acc"})

    fixmatch_cifar10 = pd.read_csv(f"FixMatch+CIFAR10+ratio+{ratio}.csv").iloc[:,:2]
    fixmatch_cifar10 = fixmatch_cifar10.rename(columns={fixmatch_cifar10.columns[1]:"unlabeled_train_acc"})

    plt.plot(fixmatch_cifar10['Step'],fixmatch_cifar10['unlabeled_train_acc'], label='FixMatch')
    plt.plot(calimatch_cifar10['Step'],calimatch_cifar10['unlabeled_train_acc'], label='Proposed (FixMatch+AcatS)')
    plt.legend()
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.show()
```
